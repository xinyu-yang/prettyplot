```python
import numpy as np
import matplotlib.pyplot as plt
import prettyplot as ppt
from prettyplot.config import *
```

```python
fig_setting = {                                                            
                "title": "The title of the figure",
                "scale": (0.5, 0.5), # The scale to expand, which is used to set the figure size
                "title_size": 16,
                "label_size":14, # The size of x and y labels
                "line_width": 2,
                "tick_labelsize": 12, # x, y tick label size
                # "xlim": [0, 10],  # xlim and ylim
                # "ylim": [0, 1],
                "xlabel": "The label of x axis",
                "ylabel": "The label of y axis",
                "bar_edgecolor": "white",
                "bar_width": 1,  # the width of the bars
                "group_padding": 2, # the width between groups
              }                                                            
settings = [                                                               
    {
        # https://github.com/matplotlib/cheatsheets
        "color": GNU_COLOR[0], # Other colormaps: plt.get_cmap("Dark2").colors[0]
        "label": "sin(x)",
        "linestyle": "-",
        "marker": MARKERSTYLE[0],
    },                                                                     
    {                                                                      
        "color": GNU_COLOR[1],                                             
        "label": "cos(x)",                                                 
        "linestyle": "--",                                                 
        "marker": MARKERSTYLE[1],                                          
    },                                                                     
] 
```

```python
x = np.linspace(0, 2*np.pi, 20)                                            
y1 = np.sin(x)                                                             
y2 = np.cos(x)                                                             
                                                                           
fig, ax = ppt.line.lines([x, x], [y1, y2], settings, fig_setting)          

# Set the border the figure
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# ax.set_yticks(np.arange(0, 75, 20).tolist())
# ax.set_yscale("log", base=2)

# Set the background color of the image
fig.patch.set_visible(False)
ax.patch.set_visible(False)

# The legend
ax.legend(frameon=False, fontsize=12)

# Plot grid
# plt.grid(linestyle="--")

# Show
plt.show()

# save figure
# fig.savefig("test.pdf", transparent=True, bbox_inches='tight')
```

```python
from random import randint

score = range(100)
label1 = [0 if i<20 else 1 for i in score[:50]]
label1.extend([1 if i<80 else 0 for i in score[50:]])
label2 = [0 if i<50 else 1 for i in score]

roc_settings = [                                                               
    {                                                                      
        "color": GNU_COLOR[0],                                                                                             
        "linestyle": "-",                                                  
        "marker": MARKERSTYLE[0],                                          
    },                                                                     
    {                                                                      
        "color": GNU_COLOR[1],                                                                                            
        "linestyle": "--",                                                 
        "marker": MARKERSTYLE[1],                                          
    },                                                                     
]  

fig, ax = ppt.roc.rocs([label1, label2], [score, score], roc_settings, fig_setting)
# The legend
ax.legend(frameon=True, fontsize=12, loc="lower right")
plt.show()
```

```python
fig_setting = {                                                            
                "title": "The title of bars",
                "scale": (0.7, 0.5), # The scale to expand, which is used to set the figure size
                "title_size": 15,
                "label_size":12, # The size of x and y labels
                "line_width": 2,
                "tick_labelsize": 12, # x, y tick label size
                # "xlim": [0, 10],  # xlim and ylim
                # "ylim": [0, 1],
                # "xlabel": "The label of x axis",
                "ylabel": "Length (cm)",
    
                # For bar only.
                "bar_edgecolor": "white",
                "bar_width": 1,  # the width of the bars
                "group_padding": 3, # the width between groups
                "hatch": None, # The hatch to fill the bar, {'/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*'}.
                "bar_colors": plt.get_cmap("Dark2").colors, # Colors of bars, this is a list.
                # Labels of default entities and attributes, they should be matched with default data_matrix.
                # e.g., the shape of data_matrix is (3x5), then len(labels) and len(attributes) must be 3 and 5.
                "labels": ("Alice", "Bob", "Eve"), 
                "attributes": ("Height", "Shoulder Breadth", "Chest Circumference"),
                "bar_inverse_data_matrix": False, # whether to inverse the data matrix of bars, see bar.py
                "bar_label": True, # Whether to put values in front of the bars.
                "bar_label_size": "small", # ["small", "medium", "large"]
              }

data_mtx = [
    [170, 50, 100],
    [175, 65, 80],
    [180, 70, 90]
]

fig, ax = ppt.bar.bars(data_mtx, fig_setting)

# Set the border the figure
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# Set the background color of the image
fig.patch.set_visible(False)
ax.patch.set_visible(False)

ax.legend(frameon=False, fontsize=12)

plt.show()

# save figure
# fig.savefig("shape_info.png", transparent=True, bbox_inches='tight')
```
